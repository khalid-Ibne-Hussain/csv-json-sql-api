**Top Rated Movie Data import through API from TMDB
and convert dataframe JSON to CSV**


```python
import pandas as pd
import requests
```

```python
response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=d9385c9b80dde6d254b866288f2585c0&language=en-US&page=1')
```

```python
response.json()
```

```python
df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
```

```python
# Empty DataFrame 
# df= pd.DataFrame()

df_new=pd.DataFrame()
df_new
```

```python
for i in range(1,499):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=d9385c9b80dde6d254b866288f2585c0&language=en-US&page={}'.format(i))
    temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
    df_new = df_new.append(temp_df,ignore_index=True)
    
```
